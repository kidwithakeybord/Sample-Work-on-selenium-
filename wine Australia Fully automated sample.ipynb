```python
def select_year(Y):
    while True:
        try:
            driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[9].click()
            sleep(1.5)
            try:
                Year=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[9].find_elements(By.CSS_SELECTOR,'div.row')[Y]
                driver.execute_script("arguments[0].scrollIntoView();", Year)
                sleep(2)
                year=Year.text
                print(year)
                Year.click()
            except:
                driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[9].click()
                sleep(1.5)
                Year=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[9].find_elements(By.CSS_SELECTOR,'div.row')[Y]
                driver.execute_script("arguments[0].scrollIntoView();", Year)
                sleep(2)
                year=Year.text
                print(year)
                Year.click()
            break
        except:
            print("Year select error page reload")
            set_page()
    sleep(2)
    return year

def get_mount_list(driver,Y):
    #month
    #get mounth list
    kill_while_loop = 0
    while True:
        try:
            if kill_while_loop < 2:
                driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                sleep(1.5)
                try:
                    Monthlist=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'div.row')
                    title=[i.get_attribute('title') for i in driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'span.slicerText')]
                    driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                    sleep(1.5)
                except:
                    driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                    sleep(1.5)
                    Monthlist=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'div.row')
                    title=[i.get_attribute('title') for i in driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'span.slicerText')]
                    driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                    sleep(1.5)
                break
            else:
                break
        except:
            print('Mount list error reload')
            kill_while_loop =kill_while_loop+1
            set_page()
            select_year(Y)
    print(title)
    return Monthlist

def select_month(M,Y):
    kill_while_loop = 0
    while True:
        try:
            if kill_while_loop < 2:
                driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                sleep(1.5)
                try:
                    Month=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'div.row')[M]
                    driver.execute_script("arguments[0].scrollIntoView();",Month)
                    month=Month.text
                    print(month)
                    sleep(2)
                    Month.click()
                except:
                    driver.find_elements(By.CSS_SELECTOR,'div.slicer-dropdown-menu')[8].click()
                    sleep(1.5)
                    Month=driver.find_elements(By.CSS_SELECTOR,'div.visibleGroup')[8].find_elements(By.CSS_SELECTOR,'div.row')[M]
                    driver.execute_script("arguments[0].scrollIntoView();",Month)
                    month=Month.text
                    print(month)
                    sleep(2)
                    Month.click()
                break
            else:
                break
        except:
            print("Mount select error page reload")
            kill_while_loop =kill_while_loop+1
            set_page()
            select_year(Y)
    sleep(2)
    return month
```
